```python
import cv2
import os
import random
import matplotlib.pyplot as plt
from paddleocr import PaddleOCR
```

```python
ocr = PaddleOCR(det_model_dir='ppocr/ch_PP-OCRv3_det_infer',
                rec_model_dir='ppocr/ch_PP-OCRv3_rec_infer',
                cls_model_dir='ppocr/ch_ppocr_mobile_v2.0_cls_infer', 
                use_angle_cls=True)
```

```python
path = 'snippets/crop'
imgs = os.listdir(path)
img = os.path.join(path,(random.choice(imgs)))
print(img)
plt.imshow(cv2.imread(img))
```

```python
result = ocr.ocr(img,det=False,rec=True,cls=True)
result
```

```python
result = ocr.ocr(img,det=True,rec=True,cls=True)
result
```
