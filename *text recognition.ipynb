# 2. curve scene text

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from utils import Extractor
```

```python
img = cv2.imread("ctr/data/test_images/arc_text.jpg")

with open('ctr/data/test_txt/arc_text.txt','r') as f:
    line = f.read()

print(line)
plt.imshow(img)
```

```python
pts = line.split(',')
print(len(pts))
```

```python
poly = []
for i in range(0,len(pts),2):
    x=pts[i]
    y=pts[i+1]
    poly.append((int(x),int(y)))
```

```python
poly
```

```python
img = cv2.circle(img,poly[0],3,(0,255,255),-1)
img = cv2.circle(img,poly[-1],3,(0,0,255),-1)
plt.imshow(img)
```

```python
# create instance of Extractor
extractor = Extractor()
# load model
extractor.load_model(det_model_dir="ppocr/ch_PP-OCRv3_det_infer",
                     rec_model_dir="ppocr/ch_PP-OCRv3_rec_infer")
```

```python
f = 'arc_text'
# read input image
image_file,test_image = extractor.read_image(f"ctr/data/test_images/{f}.jpg")
image = cv2.resize(image_file,(1280,1280))
print(image.shape)
# # find mask
_,mask = extractor.detect(image_file,test_image)
#dilate
k = 31
kernel = np.ones((k,k),np.uint8)
dilate = cv2.dilate(mask.astype('uint8')*255,kernel,iterations=1)
# display
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(13,6))
axes[0].imshow(image)
axes[1].imshow(mask)
axes[2].imshow(dilate)
# save to ctr
cv2.imwrite(f'ctr/data/test_images/{f}.jpg',image)
```

```python
contours, hierarchy = cv2.findContours(dilate, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
print("Detected: ",len(contours)," contours")

areas = []
for cnt in contours:
    area = cv2.contourArea(cnt)
    areas.append(area)

i = np.argmax(areas)
cnt = contours[i]
print(cnt.shape)
```

```python
img = image.copy()
# draw contours
cv2.drawContours(img,cnt,-1, (0,255,0), 3)
# draw bounding rectangle
x,y,w,h = cv2.boundingRect(cnt)
print(x,y,w,h)
cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),3)
# rotated rectangle
rect = cv2.minAreaRect(cnt)
box = cv2.boxPoints(rect)
box = np.int0(box)
cv2.drawContours(img,[box],0,(0,0,255),2)
plt.imshow(img)
```

```python
cnt
```

```python
pts = cnt.reshape(-1,2)
pts
```

```python
img = cv2.polylines(img, [pts],True,(255,0,255),2)
plt.imshow(img)
```

```python
print(x,y,w,h)
```

```python
cv2.norm(pts[1]-pts[0])
```

```python
pts.shape
```

```python
xmin = pts[0][0]
start_point = pts[0]
for pt in pts:
    if pt[0] < xmin:
        xmin = pt[0]
        start_point = pt
```

```python
start_point
```

```python
l = []
idx = 0
for i,pt in enumerate(pts):
    l.append(pt)
    if pt[0]==start_point[0] and pt[1]==start_point[1]:
        idx = i

print(len(l))
print('start index: ',idx)
```

```python
a = list(range(len(l)))
b = []
for i in a:
    b.append(i - idx)
b
```

```python
s= [None]*len(l)

for i in range(len(l)):
    s[b[i]] = l[i]
    
s
```

```python
p = []
D_MIN = 40
pt_0 = s[0]
# check limit
for i in range(len(s)):
    # get x,y
    pt = s[i]
    px,py = pt
    # check x
    if px < x or px > x+w:
        continue
    # check y
    elif py < y or py > y+h: 
        continue
    # check
    d = cv2.norm(pt-pt_0)
    if d > D_MIN:
        p.append(px)
        p.append(py)
        pt_0 = pt # update previous point

print(len(l))
print(len(s))
print(len(p)/2)
```

```python
m = p[:-2]
# export text
line = ""
for pt in m:
    line+=str(pt)+","

line = line[:-1]
with open('ctr/data/test_txt/demo.txt','w') as f:
    f.write(line)
```
