```python
import os
import cv2
import json
import random
import matplotlib.pyplot as plt
from utils import *
```

```python
# instance of CodeReader
reader = CodeReader()

# load aliases
with open('aliases.json','r') as f:
    aliases = json.load(f)
aliases
```

```python
path = "samples"
imgs = os.listdir(path)
```

```python
# read image
imgs = ['01.jpg','02.jpg','04.jpg','08.jpg','09.jpg','12.jpg','15.jpg','16.jpg','17.jpg','22.jpg','23.jpg','25.jpg']
img =  cv2.imread(os.path.join(path,random.choice(imgs)))
d_list = reader.read_code(img)
img = reader.draw_code(img,d_list)
# print out
for i,d in enumerate(d_list):
    print(i,d)
plt.imshow(img)
```

```python
s = []
a = []

for de in d_list:
    t = de.type
    data = de.data.decode()
    if t == "QRCODE":
        if data[0]=="1":
            data = data.split('$')
        else:
            data = data.split(',')
        for d in data:
            if d[:2]=="S:":
                s.append(d[2:])
            elif d[:2]=="I:":
                if d[2:] in aliases.keys():
                    a.append(aliases[d[2:]])
                else:
                    a.append(d[2:])
            elif d[:2]=="C:":
                if d[2:] in aliases.keys():
                    a.append(aliases[d[2:]])
                else:
                    a.append(d[2:])
    elif t == "CODE128":
        if (len(data)==14 or len(data)==16) and all(c.isalnum() for c in data):
            s.append(data)
        elif data[:3]=="SN:":
            s.append(data[3:])
    else:
        continue
    
# find serial by the most frequent
serial = None
if len(s) != 0:
    count = 0
    serial = s[0]
    for i in s:
        f = s.count(i)
        if f > count:
            count = f
            serial = i
            
# find model by the most frequent
model = None
if len(a) != 0:
    count = 0
    model = a[0]
    for i in a:
        f = a.count(i)
        if f > count:
            count = f
            model = i
    
print("SERIAL: ",serial)
print("MODEL: ",model)
```

```python
path = "samples"
imgs = ['01.jpg','02.jpg','04.jpg','08.jpg','09.jpg','12.jpg','15.jpg','16.jpg','17.jpg','22.jpg','23.jpg','25.jpg']

# instance of CodeReader
reader = CodeReader()

# load aliases
with open('aliases.json','r') as f:
    aliases = json.load(f)
    
# read image
img =  cv2.imread(os.path.join(path,random.choice(imgs)))

model,serial = reader.infer(img,aliases)

# draw
d_list = reader.read_code(img)
img = reader.draw_code(img,d_list)

plt.imshow(img)
plt.title(f"M:{model} S:{serial}")
```
