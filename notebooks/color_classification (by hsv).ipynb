**create table**

```python
import os
from PIL import Image
import random
import matplotlib.pyplot as plt
```

```python
path = "gg"
labels = os.listdir(path)
imgs = []
for label in labels:
    f = random.choice(os.listdir(os.path.join(path,label)))
    img = Image.open(os.path.join(path,label,f))
    imgs.append(img)
```

```python
cols = 3 # odd
plt.figure(figsize=(40,20))
for i in range(len(imgs)):
    plt.subplot(len(imgs)%cols+1,cols,i+1)
    plt.imshow(imgs[i])
    plt.title(labels[i])
    
plt.savefig("table.jpg",dpi=200)
```
