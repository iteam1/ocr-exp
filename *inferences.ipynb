```python
for r in res[0]:
    t = r[1][0].lower()
    print(t)
```

```python
model = None
serial = None
for r in res[0]:
    t = r[1][0].lower()
    print(t)
    # infer model
    count = t.count('h2c')
    if count > 2:
        model = 'H2C'
    # infer serial 14 digit
    if len(t)>14:
        for i in range(len(t)):
            s = t[i:i+14]
            if len(s) == 14\
                and all([c.isalnum() for c in s])\
                and all([c.isnumeric() for c in s[:4]]):
                print('serial: ',s)
                serial = s
        
print(model.upper(),serial.upper())
```
