# 3d visualization text

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# setting up the figure and axes
fig = plt.figure(figsize=(7,4))
ax = fig.add_subplot(111,projection='3d')

# arranging data
_x = np.arange(5)
_y = np.arange(4)
_xx, _yy = np.meshgrid(_x,_y)
x,y = _xx.ravel(),_yy.ravel()

top = x+y
bottom = np.zeros_like(top)
width = depth = 1

#plotting the bard with shade
ax.bar3d(x,y,bottom,width,depth,top,shade=True,color='salmon')
ax.set_title('Shaded 3D bar plot')
plt.show()
```
