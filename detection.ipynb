```python
import os
```

```python
train_images = os.listdir('dataset/ch4_training_images')
test_images = os.listdir('dataset/ch4_test_images')
train_labels = os.listdir('dataset/ch4_training_localization_transcription_gt')
test_labels = os.listdir('dataset/Challenge4_Test_Task1_GT')

print(f'train images: {len(train_images)}')
print(f'test images: {len(test_images)}')
print(f'train labels: {len(train_labels)}')
print(f'test images: {len(test_labels)}')
```

```python
import json
```

```python
# train set
train_label = open("dataset/train_label.txt","w")
train_label_dir = 'dataset/ch4_training_localization_transcription_gt'
train_image_dir = 'dataset/ch4_training_images'

for file in train_labels:

    # file = train_labels[0]
    img_name = 'img_'+ file.split('.')[0].split('_')[-1] + '.jpg'

    with open(os.path.join(train_label_dir,file),'r') as f:
        content = f.readlines()

    content = [x.strip() for x in content]

    text = []

    for i in content:

        label = {}

        i = i.replace('\ufeff','')

        i = i.split(',')

        label['transcription'] = i[-1]

        label['points'] = [[i[0],i[1]],[i[2],i[3]],[i[4],i[5]],[i[6],i[7]]]

        text.append(label) #print(label)

    content = []
    text = json.dumps(text, ensure_ascii=False)

    train_label.write( img_name+ '\t'+f'{text}' + '\n')

train_label.close()
```

```python
# train set
test_label = open("dataset/test_label.txt","w")
test_label_dir = 'dataset/Challenge4_Test_Task1_GT'
test_image_dir = 'dataset/ch4_test_images'

for file in test_labels:

    img_name = 'img_'+ file.split('.')[0].split('_')[-1] + '.jpg'

    with open(os.path.join(test_label_dir,file),'r') as f:
        content = f.readlines()

    content = [x.strip() for x in content]

    text = []

    for i in content:

        label = {}

        i = i.replace('\ufeff','')

        i = i.split(',')

        label['transcription'] = i[-1]

        label['points'] = [[i[0],i[1]],[i[2],i[3]],[i[4],i[5]],[i[6],i[7]]]

        text.append(label) #print(label)

    content = []
    text = json.dumps(text, ensure_ascii=False)

    test_label.write( img_name+ '\t'+f'{text}' + '\n')

test_label.close()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
# When using matplotlib.pyplot to draw in the notebook, you need to add this command to display
%matplotlib inline
from PIL import Image
import numpy as np
```

```python
with open('dataset/train_label.txt','r') as f:
    lines = f.readlines()
    
lines
```

```python
# 2. Fetch the first data
line = lines[0]
line
```

```python
img_name, gt_label = line.strip().split("\t")
print("img_name: ",img_name)
print("gt_label: ",gt_label)
```

```python
# 3. Read image
image = open(os.path.join(train_image_dir, img_name), 'rb').read()
data = {'image': image, 'label': gt_label}
```

```python
# https://github.com/PaddlePaddle/PaddleOCR/blob/release%2F2.4/ppocr/data/imaug/operators.py
from ppocr.data.imaug.operators import DecodeImage
```

```python
# 4. Declare the DecodeImage class to decode the image
decode_image = DecodeImage(img_mode='RGB', channel_first=False)
data = decode_image(data)

# 5. Print the shape of the decoded image and visualize the image
print("The shape of decoded image is ", data['image'].shape)

plt.figure(figsize=(10, 10))
plt.imshow(data['image'])
src_img = data['image']
```

```python
# Import DetLabelEncode from PaddleOCR
from ppocr.data.imaug.label_ops import DetLabelEncode

# 1. Declare the class for label decoding
decode_label = DetLabelEncode()

# 2. Print the label before decoding
print("The label before decode are: ", data['label'])

# 3. Label decoding
data = decode_label(data)
print("\n")

# 4. Print the decoded label
print("The polygon after decode are: ", data['polys'])
print("The text after decode are: ", data['texts'])
```
